## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df_ConvertedComp = df['ConvertedComp']

df_ConvertedComp.plot(kind = 'box')

plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and 'ConvertedComp' is the column of interest
df_ConvertedComp = df['ConvertedComp']

# Create a box plot
sns.boxplot(x=df_ConvertedComp)

# Calculate and print the interquartile range (IQR)
Q1 = df['ConvertedComp'].quantile(.25)
Q3 = df['ConvertedComp'].quantile(.75)
IQR = Q3 - Q1
print("Interquartile Range (IQR):", IQR)

# Show the plot
plt.show()
```

Find out the upper and lower bounds.


```python
# your code goes here
upper = Q3 + (IQR * 1.5)
lower = Q1 - (IQR * 1.5)

print(f"Upper Bond: {upper}")
print(f"Lower Bond: {lower}")
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
(df['ConvertedComp']<lower) | (df['ConvertedComp']>upper)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df2 = df['ConvertedComp'].clip(upper, lower)
df2.describe()
```
